#### What is Seaborn?
##### Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics with less code. Seaborn makes it easier to generate complex plots like: heatmaps, violin plots, pair plots and more often with just a single function call.

#### 2. Distribution Plot
##### A distribution plot shows how our data values are spread out â€” whether they are concentrated around a mean, skewed, or uniformly spread.
##### Plot under distribution plots
1) histplot()
2) kdeplot()
3) rugplot()


#### Figure level function : 
##### displot()
#### Axes level function:
##### histplot() , kdeplot() , rugplot()

#### a) Plotting univariate histogram

```python
sns.histplot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
plt.show()
```

##### Normally, a histogram is used for numerical data â€” it divides numbers into bins (ranges) like 0â€“10, 10â€“20, etc. But if our data is categorical (like Male/Female, Red/Blue/Green, or Yes/No), Seaborn/Matplotlib can still make a bar-like plot showing how many times each category appears.

```python
sns.displot(data=tips,x='day',kind='hist',hue='sex')
plt.show()
```

#### **element** : 
##### element parameter controls how the bars of the histogram are drawn â€” basically, the style or look of each bar (especially when we use hue)
##### Values:
1) 'bars'
2) 'step'
3) 'ploy'

```python
sns.displot(data=tips,x='day',kind='hist',hue='sex',element='step')
plt.show()
```

##### faceting using col and row

```python
sns.displot(data=tips, x="total_bill", col="sex", row='day',hue='smoker', kind='hist')
plt.show()
```

#### b) kdeplot() -> (Kernel Density Estimation)
##### It estimates how our data is spread out and draws a continuous smooth line instead of using bars (like a histogram).
##### For Example: The probability that the bill will be $15.

```python
sns.displot(data=tips,x='total_bill',kind='kde')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',fill=True)
plt.show()
```

#### rugplot()
##### rugplot() shows small vertical lines (ticks) along an axis to represent each individual data point in your dataset. Itâ€™s like placing a small mark for every data value, helping you see how data points are spread out.
##### (It is like histogram)

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
plt.show()
```

##### Bivariate histograms
A bivariate histogram shows how two numerical variables are related by dividing the plot area into small rectangles (bins).
Each rectangleâ€™s color shows how many data points fall into that area.

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

##### Bivariate kdeplot
A bivariate KDE plot shows the density of two variables (x and y) as a smooth surface or contour, instead of using rectangles.
Itâ€™s like a 2D version of a KDE plot, where the â€œheightâ€ or â€œcolor intensityâ€ tells you where data points are concentrated.
(Line gap -> Less Number)

```python
sns.kdeplot(data=tips,x='total_bill',y='tip')
plt.show()
```

#### a) heatmap()
A heatmap is a graphical representation of data where individual values in a matrix (2D array, DataFrame, or correlation table) are shown as colors.
Plot rectangular data as colored_encoded matrix

```python
temp_df=gap[gap['continent']=='Asia'].pivot_table(index='country',columns='year',values='lifeExp')
temp_df
```

#### b) clustermap() : 
clustermap() similar to heatmap(), but it adds hierarchical clustering to automatically group similar rows and columns based on their data values.

#### 4) Categorical Plots
Categorical data refers to data that represents discrete groups or categories (like gender, color, or type) rather than numerical values.

#### Types
1) Categorical Scatter Plots
##### i) stripplot()
##### ii) swarmplot()
2) Categorical Distribution Plots
##### i) Boxplot()
##### ii) Violinplot()
3) Categorical Estimate Plot
##### i) Barplot()
##### ii) Pointplot()
##### iii) Countplot()

#### Figure level function is catplot()

#### 1) Categorical Scatter Plots

##### i) striplot() -> Plots individual data points for categorical variables using a scatter plot.

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

##### Seaborn adds a slight random noise (called jitter) to spread out data points for better visibility. To align all points on a single line (remove the noise), set the parameter jitter=False.

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',jitter=False)
plt.show()
```

##### Figure level function -> catplot()

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',hue='sex',jitter=False)
plt.show()
```

##### ii) swarmplot() -> Similar to stripplot but adjusts points to avoid overlap.
##### This is better for small data, stripplot() is mostly used for medium and large data

```python
sns.swarmplot(data=tips,x='sex',y='tip',hue='day')
plt.show()
```

#### 2) Categorical Distribution Plots

##### i) Boxplot -> A boxplot is a standardized way to display the distribution of data using a five-number summary (minimum, first quartile, median, third quartile, and maximum). It helps identify outliers, shows data symmetry, reveals how tightly data is grouped, and indicates whether the data is skewed.


```python
sns.boxplot(data=tips,x='day',y='total_bill')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='box')
plt.show()
```

##### single boxplot -> numerical column

```python
sns.catplot(data=tips,x='tip',kind='box',hue='day')
plt.show()
```

##### ii) Violinplot() -> Combines a boxplot with a kernel density plot to show data distribution shape.

```python
sns.violinplot(data=tips,y='total_bill',x='day',hue='sex')
plt.show()
```

```python
sns.catplot(data=tips,y='total_bill',x='day',hue='sex',split=True,kind='violin')
plt.show()
```

#### 3) Categorical Estimate Plot

##### i) barplot() â€“ Displays the mean (or other estimator) of a numerical variable for each category.

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
# This is showing average bill by male and female in different day. i.e estimator=mean
```

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',errorbar=None)
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='bar',errorbar=None,estimator=np.max)
plt.show()
```

##### ii) pointbar() -> Plots point estimates with confidence intervals for each category.
##### This gives the difference (connects the bar)

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='point',hue='sex',errorbar=None)
plt.show()
```

##### countplot() -> Shows the count of observations in each categorical bin.
##### This is like values_count().

```python
sns.catplot(data=tips,x='sex',kind='count',hue='day')
plt.show()
```

#### faceting using catplot()

```python
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',row='time',kind='bar',hue='day',errorbar=None)
plt.show()
```

#### Regression: A statistical method that shows the relationship between two quantitative variables.

#### Types:

##### regplot() -> An axes-level function for plotting data and a regression line on a single subplot.

##### lmplot() -> A figure-level function that can create multiple regression plots with additional facets.
##### residplot() -> An axes-level function that plots the residuals of a regression model to check the fit and detect patterns or deviations.


##### i) regplot()

```python
sns.regplot(data=tips,x='tip',y='total_bill')
plt.show()
```

##### ii) lmplot()

```python
sns.lmplot(data=tips,x='tip',y='total_bill',hue='sex')
plt.show()
```

##### iii) residplot()

```python
sns.residplot(data=tips,y='tip',x='total_bill')
plt.show()
```

#### Plotting Pairwise Relationship (PairGrid vs Pairplot)

#### Pairplot : 
##### automatically detect the numerical column
##### pair-by-pair plot scatter-plot
##### If same then histogram

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

##### pairplot is build by PairGrid internally

```python
g=sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g=sns.PairGrid(data=iris)
g.map_diag(sns.scatterplot)
g.map_offdiag(sns.histplot)
```

```python
g=sns.PairGrid(data=iris)
g.map_diag(sns.scatterplot)
g.map_upper(sns.histplot)
g.map_lower(sns.barplot)
```

```python
# Restriction 
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
```
