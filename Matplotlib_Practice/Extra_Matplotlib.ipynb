### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv("iris.csv")
iris.sample(5)
```

#### Scatter

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('petal Length')
plt.show()
```

```python
# Renaming the Species
# 0 -> Iris-virginica
# 1-> Iris-setosa
# 2 -> Iris-versicolor
iris['Species']=iris['Species'].replace(['Iris-virginica','Iris-setosa','Iris-versicolor'],[0,1,2])
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=0.7,cmap='jet')
plt.colorbar()
plt.show()
# Which color represent what -> colorbar()
```

#### figure()
#### To change the size of graph

```python
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=0.7,cmap='jet')
plt.colorbar()
plt.show()
```

```python
batter=pd.read_csv('batter.csv')
temp_df=batter.head(100)
temp_df=temp_df.sample(25)
temp_df
```

#### Annotations
##### text() function is used to add text annotations at an arbitrary location on the plot.
##### It allows us to place any text at a specified (x, y) coordinate inside the plotting area. 
##### This is different from title(), xlabel(), or ylabel(), which place text in specific predefined positions.

```python
plt.scatter(temp_df['avg'],temp_df['strike_rate'])
for i in range(temp_df.shape[0]):
    plt.text(temp_df['avg'].values[i],temp_df['strike_rate'].values[i],temp_df['batter'].values[i])
```

#### Horizontal and Vertical Lines
##### Player having average more than 30 and strike rate more than 140

```python
plt.scatter(temp_df['avg'],temp_df['strike_rate'])
for i in range(temp_df.shape[0]):
    plt.text(temp_df['avg'].values[i],temp_df['strike_rate'].values[i],temp_df['batter'].values[i])
plt.axhline(130,color='red')
plt.axvline(30,color='red')
```

#### Subplots

```python
fig,ax=plt.subplots()
# fig -> Figure Object  ax -> axis object
# Instead of plt.title() we use ax.set_title() (All functions starts with set_)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,sharex=True,figsize=(10,6))
ax[0][0].scatter(batter['avg'],batter['runs'])
ax[0][0].set_title("Average and Runs")
ax[0][0].set_ylabel("Runs")

ax[1][0].scatter(batter['avg'],batter['strike_rate'])
ax[1][0].set_title("Average and Strike Rate")
ax[1][0].set_ylabel("Strike Rate")
ax[1][0].set_xlabel("Average")

ax[0][1].bar(batter['avg'],batter['runs'])
ax[0][1].set_title("Average and Runs")
ax[0][1].set_ylabel("Runs")

ax[1][1].bar(batter['avg'],batter['strike_rate'])
ax[1][1].set_title("Average and Strike Rate")
ax[1][1].set_ylabel("Strike Rate")
ax[1][1].set_xlabel("Average")

```

#### Other ways for subplots

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1) # -> 2 rows, 2 columns, 1st position
ax1.scatter(temp_df['avg'],temp_df['strike_rate'],color='red')
ax2=fig.add_subplot(2,2,2)
ax2.scatter(temp_df['avg'],temp_df['strike_rate'],color='blue')
ax3=fig.add_subplot(2,2,3)
ax3.hist(temp_df['avg'])
ax4=fig.add_subplot(2,2,4)
ax4.bar(temp_df['avg'],temp_df['strike_rate'],color='green')
plt.show()
```

### 3D Plots

#### 3D Scatter Plots

```python
# fig=plt.figure()
# fig object contain main graph
ax=plt.subplot(projection='3d')
ax.scatter3D(batter['runs'],batter['avg'],batter['strike_rate'])
plt.show()
```

#### 3D Line Plot

```python
x=[0,5,-4]
y=[1,5,-10]
z=[6,-8,11]
ax=plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100])
plt.show()
```

#### 3D Surface Plots

```python
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='plasma')
plt.colorbar(p)
```

#### Other function

```python
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)
xx,yy=np.meshgrid(x,y)
z=np.sin(xx)+np.cos(yy)
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='plasma')
plt.colorbar(p)
```

#### Contour Plot 
#### 3D convert into 2D

```python
plt.figure(figsize=(10,10))
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
ax=plt.subplot(projection='3d')
p=ax.contourf(xx,yy,z,cmap='plasma')
plt.colorbar(p)
```

#### Heatmap

```python
delivary=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
```

### Generate a heatmap that show 6's hit on every ball of every over

```python
plt.figure(figsize=(15,7))
temp_delivery=delivary[(delivary['ballnumber'].isin([1,2,3,4,5,6])) & (delivary['batsman_run']==6)]
grid=temp_delivery.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
plt.imshow(grid, aspect='auto', cmap='viridis')
plt.colorbar()
plt.xlabel('Ball Number')
plt.ylabel('Overs')
plt.xticks(ticks=range(grid.shape[1]), labels=grid.columns)
plt.yticks(ticks=range(grid.shape[0]), labels=(grid.index)+1)
plt.show()
```

#### generate a heatmap that shows wicket on every ball of every overs

```python
plt.figure(figsize=(15,7))
temp_delivery=delivary[(delivary['ballnumber'].isin([1,2,3,4,5,6])) & (delivary['isWicketDelivery']==1)]
grid=temp_delivery.pivot_table(index='overs',columns='ballnumber',values='isWicketDelivery',aggfunc='count')
plt.imshow(grid, aspect='auto', cmap='plasma')
plt.colorbar()
plt.xticks(ticks=range(grid.shape[1]),labels=grid.columns)
plt.yticks(ticks=range(grid.shape[0]),labels=(grid.index)+1)
plt.show()
```
