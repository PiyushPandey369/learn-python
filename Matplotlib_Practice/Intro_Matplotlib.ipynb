### from pandas dataframe

```python
batsman=pd.read_csv("sharma-kohli.csv")
plt.plot(batsman["index"],batsman["V Kohli"])
```

### Multiple plots

```python
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

### labels title and parameters of plot()

```python
plt.title('Rohit vs Virat')
plt.xlabel("Season")
plt.ylabel("Runs_Scored")
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

### Color,line,marker,label

```python
plt.plot(batsman["index"],batsman["V Kohli"],color='red',linestyle='dashed',linewidth=3,marker='o',markersize=10,label="Virat Kholi")
plt.plot(batsman["index"],batsman["RG Sharma"],color='black',linestyle='dotted',linewidth=3,marker='D',markersize=10,label="Rohit Sharma")
plt.title('Rohit vs Virat')
plt.xlabel("Season")
plt.ylabel("Runs_Scored")
plt.legend(loc='best')
plt.show()
```

### grid and show

```python
plt.plot(batsman["index"],batsman["V Kohli"],color='red',linestyle='dashed',linewidth=3,marker='o',markersize=10,label="Virat Kholi")
plt.plot(batsman["index"],batsman["RG Sharma"],color='black',linestyle='dotted',linewidth=3,marker='D',markersize=10,label="Rohit Sharma")
plt.title('Rohit vs Virat')
plt.xlabel("Season")
plt.ylabel("Runs_Scored")
plt.legend(loc='best')
plt.grid()
plt.show() # Without show() figures will not be displayed in other editor(like: VS Code)
```

### Bar Charts

#### simple bar-charts

```python
no_of_students=[20,35,39,15,10,38]
subject=['Social Studies','English','Nepali','Maths','Opt.Maths','Science']
plt.bar(subject,no_of_students,color=['red','black','blue','green','pink','gray','orange'],width=0.6)
plt.xticks(rotation='vertical')
```

#### Horizontal bar charts

```python
no_of_students=[20,35,39,15,10,38]
subject=['Social Studies','English','Nepali','Maths','Opt.Maths','Science']
plt.barh(subject,no_of_students,color=['red','black','blue','green','pink','gray','orange'])
```

#### Bar-charts more parameters

```python
plt.bar(subject, no_of_students, color='skyblue', edgecolor='black', hatch='//', alpha=0.8)
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Multiple bar charts
#### We cannot directlt plot multiple charts

```python
bat=pd.read_csv("batsman_season_record.csv")
bat
```

```python
plt.bar(np.arange(bat.shape[0])-0.2,bat['2015'],width=0.2,label='2015')
plt.bar(np.arange(bat.shape[0]),bat['2016'],width=0.2,label='2016')
plt.bar(np.arange(bat.shape[0])+0.2,bat['2017'],width=0.2,label='2017')
plt.xticks(np.arange(bat.shape[0]),bat['batsman'])
plt.legend()
plt.show()
```

### Stacked Bar Charts

```python
plt.bar(bat['batsman'],bat['2015'],label='2015')
plt.bar(bat['batsman'],bat['2016'],bottom=bat['2015'],label='2016')
plt.bar(bat['batsman'],bat['2017'],bottom=(bat['2015']+bat['2016']),label='2017')
plt.legend()
plt.show()
```
